```python
#Data Preprocessing

#importing libraries
import sklearn
import numpy as np
import pandas as pd
#importing dataset
data=pd.read_csv('Data.csv')
print("Dataset imported")
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
#identifying total number of missing values in each column
df.isnull().sum()
```

```python
#removing those null values
#There are 2 ways to do this
#1. dropna()
#2.fillna(with some value)

#1. using dropna()
df1=df.copy()
df1=df.dropna()
print("Using dropna()")
print(df1)
#2.using fillna()
df2=df.copy()
df2.fillna(df2.mean(numeric_only=True),inplace=True)
print("using fillna()")
print(df2)
df2.isnull().sum()
```

```python
#Next step is encoding
# 2 types of encoding
#Label Encoding, Onehot encoder

#The below code is for onehot encoder
df_encoded=pd.get_dummies(df)
print(df_encoded)
```

```python
#split data
from sklearn.model_selection import train_test_split
X=df[['Age', 'Salary', 'Country']]
y=df[['Purchased']]
#70% ttraining, 30% testing
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
print("X_train")
print(X_train)
print("X_test")
print(X_test)
print("y_train")
print(y_train)
print("y_test")
print(y_test)
```
